#### 3.5.2 实现三角函数

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt


def my_sin(x):

    return np.sin(x)

def my_cos(x):

    return np.cos(x)


x = np.linspace(-np.pi, np.pi, num=50)

y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)")
plt.legend()


plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()

```

#### 4.2.3 什么是范数

范数就是表示向量“大小”的量。人工智能中常用的有L2范数和L1范数。

L1范数与L2范数详解      

在机器学习和数学优化中，**范数（Norm）**是衡量向量大小（长度）的一种方法。常见的有 **L1 范数**和 **L2 范数**，它们在正则化、稀疏表示、特征选择等任务中具有重要作用。

---

📌 L1 范数（曼哈顿距离）        

L1范数也称为**曼哈顿距离（Manhattan Distance）**或**稀疏范数（Sparse Norm）**。

对一个向量 $\mathbf{x} = [x_1, x_2, \dots, x_n]$，其 L1 范数定义如下：

$
\|\mathbf{x}\|_1 = \sum_{i=1}^{n} |x_i|
$

✅ 特点：       

- 可以产生稀疏解（很多元素为0），用于特征选择（如Lasso回归）。      
- 对异常值更鲁棒。          

🧠 举例：

$
\mathbf{x} = [3, -4, 0, 5] \Rightarrow \|\mathbf{x}\|_1 = |3| + |-4| + |0| + |5| = 12
$

---

📌 L2 范数（欧几里得距离）      

L2范数也称为**欧几里得范数（Euclidean Norm）**，是我们最常用的距离度量方式。        

定义如下：      

$
\|\mathbf{x}\|_2 = \sqrt{\sum_{i=1}^{n} x_i^2}
$

✅ 特点：       

- 解更平滑，不一定稀疏，但对噪声敏感。      
- 常用于岭回归（Ridge Regression）。        

 🧠 举例：      

$
\mathbf{x} = [3, -4, 0, 5] \Rightarrow \|\mathbf{x}\|_2 = \sqrt{3^2 + (-4)^2 + 0^2 + 5^2} = \sqrt{50} \approx 7.07
$

---

📊 主要区别对比        

| 特性            | L1 范数                          | L2 范数                         |
|-----------------|----------------------------------|----------------------------------|
| 公式            | $\sum |x_i|$                    | $\sqrt{\sum x_i^2}$             |
| 结果稀疏性      | 高（适合特征选择）              | 低（结果更平滑）               |
| 对异常值敏感性  | 低                              | 高                              |
| 常见应用        | Lasso回归、稀疏编码              | 岭回归、最小二乘               |

---

✅ 总结     

- L1范数：适合需要稀疏解的场景，如变量选择；        
- L2范数：适合需要整体稳定性的场景，如模型平滑优化；        

在实际机器学习中，很多模型（如ElasticNet）也会同时使用 L1 和 L2 范数的组合。
