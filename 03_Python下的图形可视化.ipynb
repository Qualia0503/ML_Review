```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x - 2) * x * (x + 2)
```

#### 3.1.6装饰图形

```python
def f2(x, w):
    return (x - w) * x * (x + 2)

# 定义x
x = np.linspace(-3, 3, 100)

# 绘制图形
plt.plot(x, f2(x, 2), color='black', 
         label='$w=2$')
plt.plot(x, f2(x, 1), color='cornflowerblue', 
         label='$w=1$')
plt.legend(loc="upper left") # 显示图例
plt.ylim(-15, 15)            # y轴范围
plt.title('$f_2(x)$')        # 标题
plt.xlabel('$x$')            # x轴标签
plt.ylabel('$y$')            # y轴标签
plt.grid(True)               # 显示网格
plt.show()
```

#### 3.1.7 并列显示多张图形

```python
plt.figure(figsize=(10, 3))                 # 指定figure的大小

plt.subplots_adjust(wspace=0.5, hspace=0.5) # 指定子图的间距

for i in range(6):                          # 画6个图
    plt.subplot(2, 3, i + 1)
    plt.title(i + 1)
    plt.plot(x, f2(x, i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

### 3.2 绘制三维图形

#### 3.2.1 包含两个变量的函数

#### $$ f(x_{0}, x_{1}) = (2x_{0}^2 + x_{1}^2)\exp(-(2x_{0}^2 + x_{1}^2)) $$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f3(x0, x1):
    ans = (2 * x0**2 + x1**2) * np.exp(-(2 * x0**2 + x1**2))
    return ans

xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
```

```python
plt.figure(figsize=(3.5, 3))
plt.gray()
plt.pcolor(y)
plt.colorbar()
plt.show()
```

#### 3.2.3绘制三维图形:surface

```python
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, 
                color='blue', 
                edgecolor='black')
ax.set_zticks((0, 0.5))
ax.view_init(75, -95)
plt.show()
```

#### 3.2.4绘制等高线：contour

```python
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((xn, xn))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])

xx0, xx1 = np.meshgrid(x0, x1)                     # (A)

plt.figure(1, figsize=(4, 4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # (B)
cont.clabel(fmt='%.2f', fontsize=8)               # (C)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
