#### 3.1.1 绘制随机图形

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# 解决中文显示问题
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 创建数据
np.random.seed(1)
x = np.arange(10)
y = np.random.rand(10)

# 创建图形
plt.plot(x, y)
plt.show()
```
